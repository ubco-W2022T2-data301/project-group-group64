```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/val_stats.csv", low_memory = False)
```

```python
# Histogram of headshot percentage
plt.hist(df['headshot_percent'], bins = 20)
plt.title('Headshot Percentage')
plt.xlabel('Headshot Percentage')
plt.ylabel('Frequency')
plt.xlim((0, 60))
plt.show()
```

This graph is showing the headshot percentage that peaks at around 23% and ranges from 5% to 52%

This graph shows that upper ranks have a higher average headshot percentage and even higher outliers with many players having a headshot percentage above 35%

```python
# Scatter plot of headshot percentage vs. kills in radiant rank

df_radiant = df[df['rating'] == 'Radiant']
plt.scatter(df_radiant['headshot_percent'], df_radiant['kills'], s=1)
plt.xlabel('Headshot Percentage')
plt.ylabel('Kills')
plt.yscale('linear')
plt.title('Headshot Percentage vs. Kills (Radiant Rank)')
plt.show()
```

This graph is showing the amount of kills related to the headshot percentage in the top rank radiant and we can see that there is a large spread but with majority of dots being between 15-35%

```python
# Scatter plot of headshot percentage vs. wins
df_radiant = df[df['rating'] == 'Radiant']

plt.scatter(df_radiant['headshot_percent'], df_radiant['wins'], s = 1)
plt.xlabel('Headshot Percentage')
plt.ylabel('Wins')
plt.title('Headshot Percentage vs. Wins')
plt.show()
```

This graph shows wins vs headshot percentage but unfortunately this did not end up proving any correlation between the wins a player has and the rank they are in
